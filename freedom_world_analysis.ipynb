## 5.1 Top changes in Total score (last 13, 10, 5, and 3 years)

```python
# Function to find countries with the biggest change in "Total" over the last N years
def get_largest_changes(df, years_back):
    last_year = df["Year"].max()
    start_year = last_year - years_back + 1

    df_recent = df[df["Year"].between(start_year, last_year)]
    pivot = df_recent.pivot(index="Country", columns="Year", values="Total")

    # Calculate the absolute change between the first and last year of the period
    pivot["change"] = pivot[last_year] - pivot[start_year]
    result = pivot["change"].sort_values(ascending=True).head(10)
    return result

# Apply the function for the last 13, 10, 5, and 3 years
top_changes_13 = get_largest_changes(df, 13)
top_changes_10 = get_largest_changes(df, 10)
top_changes_5 = get_largest_changes(df, 5)
top_changes_3 = get_largest_changes(df, 3)


print(" Top changes in last 13 years:\n", top_changes_13)
print(" Top changes in last 10 years:\n", top_changes_10)
print("\n Top changes in last 5 years:\n", top_changes_5)
print("\n Top changes in last 3 years:\n", top_changes_3)
```

#### In Which Area Did the Most Significant Decline Occur?

Before identifying in which area Nicaragua showed the greatest deterioration, it is **necessary to normalize the results** because the **areas have different score ranges**. This happens because each area includes a different number of questions. Each question can range from 0 to 4 points, so for example, area A has three questions (A1, A2, A3), meaning its total score can range from 0 to 12.

Therefore, the first step is to create a variable *max_scores* that contains the maximum possible score for each area. Later, when we calculate the average score changes for each area, we will standardize them by dividing by the maximum possible value for that area.

```python
# Groups of sub-questions for each area
area_subquestions = {
    'A': ['A1', 'A2', 'A3'],
    'B': ['B1', 'B2', 'B3', 'B4'],
    'C': ['C1', 'C2', 'C3'],
    'D': ['D1', 'D2', 'D3', 'D4'],
    'E': ['E1', 'E2', 'E3'],
    'F': ['F1', 'F2', 'F3', 'F4'],
    'G': ['G1', 'G2', 'G3', 'G4'],
}

max_per_question = 4  # maximum score per question

# Calculate the theoretical maximum score for each area
max_scores = {area: len(questions) * max_per_question for area, questions in area_subquestions.items()}

# Display the theoretical maximum scores per area
print("Maximum theoretical scores per area:")
for area, score in max_scores.items():
    print(f"Area {area}: {score}")
```

```python
# Filter data for Nicaragua in the first and last years
df_start = df[(df['Country'] == 'Nicaragua') & (df['Year'] == 2013)]
df_end = df[(df['Country'] == 'Nicaragua') & (df['Year'] == 2025)]

# Areas of interest
areas = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

# Calculate raw difference (2025 - 2013)
raw_diff = df_end[areas].values - df_start[areas].values
raw_diff_series = pd.Series(raw_diff.flatten(), index=areas)

# Normalize the differences
normalized_diff = raw_diff_series / pd.Series(max_scores)

# Sort by actual normalized change (negative = greater deterioration)
normalized_sorted = normalized_diff.sort_values()

# Print results
print("Standardized change in each area (2025 vs. 2013):")
print(normalized_sorted)
```

Instead of comparing the values from just one year at the beginning and one year at the end, it makes more sense to compare the average values over two longer periods ( two multi-year periods) — one at the start and one at the end. This way, we get a broader view and reduce the effect of any unusual or temporary changes that might distort the comparison if we only look at a single year. (**reduce noise**)

Taking the average over multiple years helps produce a more stable and accurate picture of the real change over time.

We do it as follows:

```python
# Select data for Nicaragua
nicaragua = df[df['Country'] == 'Nicaragua']

# Areas of interest
areas = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

# Calculate average scores for each area in the two periods
initial_period = nicaragua[nicaragua['Year'].between(2013, 2016)][areas].mean()
final_period = nicaragua[nicaragua['Year'].between(2022, 2025)][areas].mean()

# Normalize averages by their maximum theoretical scores
initial_norm = initial_period / pd.Series(max_scores)
final_norm = final_period / pd.Series(max_scores)

# Calculate normalized change
differences_norm = (final_norm - initial_norm).sort_values()

print("Normalized average changes by area (2022–2025 vs. 2013–2016):")
print(differences_norm)

differences_norm.plot(kind='barh', color='red', figsize=(8,5))
plt.title('Normalized Decline in Freedom Scores by Area (Nicaragua, 2022–2025 vs. 2013–2016)')
plt.xlabel('Change in Normalized Average Score')
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('norm_decline_intervals.png')
plt.show()
```

The **normalized average changes by area from 2013–2016 to 2022–2025** show a clear pattern of **deterioration** across all areas, but to varying degrees:

- **Area D (Functioning of Government)** showed the largest decline with a normalized change of about -0.58, meaning its score dropped by 58% of its maximum possible range. This indicates a significant worsening in this area.

- **Area A (Electoral Process)** follows with a -0.50 change, a 50% drop relative to its maximum score.

- **Area E (Rule of Law)** and **Area C (Freedom of Expression and Belief)** saw moderate declines of roughly -0.33 and -0.31 respectively, indicating noticeable but less severe deteriorations.

- **Areas B (Political Pluralism and Participation)** and F **(Associational and Organizational Rights)** each declined by about -0.30, showing a similar moderate level of worsening.

- **Area G (Personal Autonomy and Individual Rights)** experienced the smallest decline at approximately -0.17, meaning it worsened the least relative to its maximum score.

Overall, all areas show deterioration, but Area D followed by Area A experienced the worst worsening, while Area G shows the least deterioration.


The idea is to compare regions based on which one experiences the largest loss relative to the maximum possible value (for example, if a question’s maximum score is 12), in order to standardize the measure. Using a simple percentage change—calculated as (Final Value − Initial Value) divided by Initial Value times 100—is not feasible here, because the initial value can be zero (for instance, when responses to certain questions are all 0, indicating no freedom). Standardizing relative to the theoretical maximum avoids this problem.

## 6.1 Hypothesis Test and Effect Size (Cohen’s d): Measuring the Democratic Decline from 2013–2015 to 2023–2025

The goal is to test whether the **observed difference in the mean democracy indicators** between the periods **2013–2015** and **2023–2025** is **statistically significant**, and to calculate **Cohen’s d** as a measure of effect size for the three indicators:

- **PR** (Political Rights)  
- **CL** (Civil Liberties)  
- **Total** (sum of PR and CL)

We perform **paired-sample hypothesis tests** because we compare the same countries over two time periods.

The hypotheses are:

$$
\begin{aligned}
H_0 &: \mu_{\text{post}} = \mu_{\text{pre}} \quad \text{(no significant change)} \\
H_1 &: \mu_{\text{post}} < \mu_{\text{pre}} \quad \text{(significant decline)}
\end{aligned}
$$

If the distribution of differences is **normal**, we use the **paired t-test**. if not, we use the **Wilcoxon signed-rank test**, a non-parametric alternative.

**Cohen’s d** measures the magnitude of the effect size and is calculated for paired samples as:

$$
d = \frac{\bar{X}_{\text{post}} - \bar{X}_{\text{pre}}}{s_d}
$$

where:

- $\bar{X}_{\text{post}}$, $\bar{X}_{\text{pre}}$ are the means of the post and pre periods, respectively  
- $s_d$ is the standard deviation of the differences

Interpretation of Cohen’s d effect size:

- 0.2 = small effect  
- 0.5 = medium effect  
- 0.8 = large effect

The objective is to demonstrate a statistically significant and substantively meaningful decline in democracy between 2013–2015 and 2023–2025, using the PR, CL, and Total indicators.

```python
import statsmodels.api as sm

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (18, 5)  # un po' più largo per 3 plot

# Filter two time periods: pre (2013-2015) and post (2023-2025)
pre_period = df[df["Year"].between(2013, 2015)]
post_period = df[df["Year"].between(2023, 2025)]

# Compute country-level averages for each period
pre_means = pre_period.groupby("Country")[["PR", "CL", "Total"]].mean().reset_index()
post_means = post_period.groupby("Country")[["PR", "CL", "Total"]].mean().reset_index()

# Merge pre and post data for comparison
df_compare = pd.merge(pre_means, post_means, on="Country", suffixes=('_pre', '_post'))

variables = ["PR", "CL", "Total"]

for var in variables:
    data_to_plot = pd.melt(
        df_compare,
        id_vars=["Country"],
        value_vars=[f"{var}_pre", f"{var}_post"],
        var_name="Period",
        value_name=var
    )

    diff = df_compare[f"{var}_post"] - df_compare[f"{var}_pre"]

    # Create a figure with 3 subplots: boxplot, QQ plot, histogram side by side
    fig, axes = plt.subplots(1, 3, figsize=(20, 5))

    # Boxplot
    sns.boxplot(x="Period", y=var, data=data_to_plot, ax=axes[0])
    axes[0].set_title(f"Distribution of {var} in 2013–2015 vs 2023–2025")
    axes[0].set_xticklabels(['2013-2015', '2023-2025'])
    axes[0].set_xlabel("Period")
    axes[0].set_ylabel(var)

    # Q-Q plot of the differences (post - pre)
    sm.qqplot(diff, line='s', ax=axes[1])
    axes[1].set_title(f"Q-Q Plot of Differences ({var} post - pre)")

    # Histogram of the differences
    sns.histplot(diff, bins=20, kde=True, ax=axes[2], color='skyblue')
    axes[2].set_title(f"Histogram of Differences ({var} post - pre)")
    axes[2].set_xlabel("Difference (post - pre)")
    axes[2].set_ylabel("Frequency")

    plt.tight_layout()
    plt.show()

# Statistical tests and Cohen's d
for var in variables:
    diff = df_compare[f"{var}_post"] - df_compare[f"{var}_pre"]
    stat_shapiro, p_shapiro = shapiro(diff)

    print(f"\nVariable: {var}")
    print(f"Shapiro-Wilk test for normality → p = {p_shapiro:.4f}")

    if p_shapiro > 0.05:
        # Normally distributed
        stat, p_val = ttest_rel(df_compare[f"{var}_pre"], df_compare[f"{var}_post"])
        test_name = "Paired t-test"
    else:
        # Not normally distributed (Wilcoxon test)
        stat, p_val = wilcoxon(df_compare[f"{var}_pre"], df_compare[f"{var}_post"])
        test_name = "Wilcoxon signed-rank test"

    mean_diff = diff.mean()

    # Cohen's d
    cohen_d = mean_diff / diff.std(ddof=1)

    print(f"{test_name}: statistic = {stat:.4f}, p-value = {p_val:.4f}")
    print(f"Mean difference (post - pre): {mean_diff:.4f}")
    print(f"Cohen's d: {cohen_d:.4f}")
```

The **Shapiro-Wilk test for normality** on the differences between the two periods (2013–2015 and 2023–2025) yielded a **p-value of 0.0000** for all three variables (PR, CL, and Total). This indicates that the differences do **not** follow a normal distribution. Therefore, we used the Wilcoxon signed-rank test, which is a non-parametric alternative suitable for non-normal data.

The **Wilcoxon signed-rank test** results show a p-value of 0.0000 (much less than the significance threshold of 0.05) for PR, CL, and Total. This means that the differences between the two periods are **statistically significant** for all three indicators, demonstrating a clear change over time.

Looking at the mean differences (post period minus pre period):

- Total: -3.6268

All values are negative, which indicates that the average scores of Political Rights (PR), Civil Liberties (CL), and their Total **have significantly decreased in the 2023–2025 period compared to 2013–2015**.

This statistically significant decline in the indicators confirms a measurable decline in democratic freedoms over the decade, as captured by these key metrics.

Furthermore, **Cohen’s d effect sizes** range between approximately -0.37 and -0.41, which suggests a **small to moderate negative effect size**. ## 6.2 Hypothesis Test (Vulnerable Countries)

#### Objective:
To determine whether countries' levels of freedom in 2013 (classified as 'Free', 'Partly Free', or 'Not Free') influenced the subsequent change in their freedom status between 2013–2015 and 2023–2025. Specifically, we aim to assess if countries that were less free in 2013 experienced a more significant decline in freedom over the following decade.

#### Hypothesis:
the level of freedom in countries, **especially the more vulnerable ones**, appears to be declining year after year. In recent years, there has, in fact, been increasing talk of a “crisis of democracy”. --> We hypothesize that countries with **lower freedom scores in 2013 have experienced a greater decline in freedom by 2025**


```python
# Difference in decline among the different groups

# Filter data for the two periods: pre (2013-2015) and post (2023-2025)
pre_period = df[df["Year"].between(2013, 2015)]
post_period = df[df["Year"].between(2023, 2025)]

# Calculate the average freedom scores ("Total") for each country in both periods
pre_means = pre_period.groupby("Country")[["Total"]].mean().reset_index().rename(columns={"Total": "Total_pre"})
post_means = post_period.groupby("Country")[["Total"]].mean().reset_index().rename(columns={"Total": "Total_post"})

# Merge the two DataFrames to compare pre and post period averages by country
df_compare = pre_means.merge(post_means, on="Country")

# Initial status of each country in 2013
df_status = df[df['Year'] == 2013][['Country', 'Status']].rename(columns={'Status': 'Initial_Status'}).drop_duplicates()

# Merge the initial status information with the comparison DataFrame
df_compare = df_compare.merge(df_status, on='Country', how='left')

# Calculate absolute and percentage differences
df_compare = df_compare[df_compare['Total_pre'] != 0].copy()

df_compare['Diff_Total'] = df_compare['Total_post'] - df_compare['Total_pre']
df_compare['Percent_Diff_Total'] = (df_compare['Diff_Total'] / df_compare['Total_pre']) * 100
```

```python
# Check normality for each group
for status in ['F', 'PF', 'NF']:
    data = df_compare[df_compare['Initial_Status'] == status]['Percent_Diff_Total']
    stat, p = shapiro(data)
    print(f'Shapiro-Wilk test for {status}: W={stat:.3f}, p={p:.3f}')

# Histogram and Q-Q plot for one group (example: 'F')
sns.histplot(df_compare[df_compare['Initial_Status'] == 'F']['Percent_Diff_Total'], kde=True)
plt.title('Histogram - group F')
plt.show()

import scipy.stats as stats
stats.probplot(df_compare[df_compare['Initial_Status'] == 'F']['Percent_Diff_Total'], dist="norm", plot=plt)
plt.title('Q-Q plot - group F')
plt.show()

# Test for homogeneity of variances (among the 3 groups)
group_F = df_compare[df_compare['Initial_Status'] == 'F']['Percent_Diff_Total']
group_PF = df_compare[df_compare['Initial_Status'] == 'PF']['Percent_Diff_Total']
group_NF = df_compare[df_compare['Initial_Status'] == 'NF']['Percent_Diff_Total']

stat, p = levene(group_F, group_PF, group_NF)
print(f"Levene test for homogeneity of variances: stat={stat:.3f}, p={p:.3f}")
```

The Shapiro-Wilk test indicates that groups F and NF significantly deviate from normality (p < 0.05), while group PF does not (p = 0.470). Additionally, Q-Q plots and histograms confirm that the distributions are not normally distributed. Since two of the three groups (F and NF) violate the normality assumption, it’s better not to use parametric tests like ANOVA. Instead, use non-parametric tests (Kruskal-Wallis)

```python
print(df_compare['Initial_Status'].value_counts())
```

The groups have different sizes, but none are too small (at least 54 observations), so the difference in sample size is not a significant issue. The Kruskal-Wallis test is appropriate in this context, and the statistical power is adequate due to the sufficient number of observations in each group.

```python
# Kruskal-Wallis test (non-parametric)

groups = [df_compare[df_compare['Initial_Status'] == status]['Percent_Diff_Total'] for status in ['F', 'PF', 'NF']]
stat, p = kruskal(*groups)
print(f"Kruskal-Wallis test: H = {stat:.4f}, p = {p:.4f}")
```

Since the p-value is much smaller than 0.05, we reject the null hypothesis that the percent changes in freedom are the same across all groups (F, PF, NF).

This indicates a statistically significant difference in freedom changes between the country groups classified as F, PF, and NF in 2013.

```python
# !pip install scikit-posthocs

# Remove rows with missing Initial_Status
df_compare_clean = df_compare.dropna(subset=['Initial_Status'])

# Prepare clean groups
groups = [df_compare_clean[df_compare_clean['Initial_Status'] == status]['Percent_Diff_Total'] for status in ['F', 'PF', 'NF']]

# Perform Kruskal-Wallis test
stat, p = kruskal(*groups)
#print(f"Kruskal-Wallis test: H = {stat:.4f}, p = {p:.4f}")

# Perform Dunn's post-hoc test with Holm correction for multiple comparisons
dunn_results = sp.posthoc_dunn(
    df_compare_clean,
    val_col='Percent_Diff_Total',
    group_col='Initial_Status',
    p_adjust='holm'
)

print(dunn_results)
```

F vs PF (p = 0.0158): Significant difference — Free countries experienced a significantly smaller percent decline in freedom compared to Partly Free countries.

F vs NF (p = 0.000088): Highly significant difference — Free countries differ greatly from Not Free countries.

PF vs NF (p = 0.1279): No significant difference — Partly Free and Not Free countries do not differ statistically.

In summary, countries that were Free in 2013 had a significantly smaller decline in freedom compared to both Partly Free and Not Free countries.

However, Partly Free and Not Free countries experienced similar declines. This suggests that weaker or unstable democracies were more vulnerable to freedom loss over time.
